```python
URL = []
for i in range(0,3):
    lnk = driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')
    for j in lnk:
        URL.append(j.get_attribute('href'))
    nxt_page = driver.find_element_by_xpath('//li[@class="a-last"]')
    time.sleep(4)
    nxt_page.click()
    time.sleep(3)
```

```python
Brand= [] 
Product_title = []
Rating = []
No_ratings = []
Price = []
Return = []
Expected_delivery = []
Availability = []
Other_details = []


for link in URL:
    
    driver.get(link)
    time.sleep(3)
    try:
        ret = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[20]/div[2]/div/div/div/div[1]/div[2]/a')
        exd = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[12]/div/div/b')
        ava = driver.find_element_by_id('availability')
        time.sleep(1)
        oth = driver.find_element_by_xpath('//ul[@class="a-unordered-list a-vertical a-spacing-mini"]')
        nrat = driver.find_element_by_id('acrCustomerReviewText')
        rat = driver.find_element_by_xpath('//span[@class="a-size-medium a-color-base"]')
        time.sleep(1)
        pri = driver.find_element_by_id('priceblock_ourprice')
        tit = driver.find_element_by_id('productTitle')
        br = driver.find_element_by_id('bylineInfo')
        time.sleep(1)
        Product_title.append(tit.text)
        Brand.append(br.text.replace('Brand:','').replace('Visit the ','').replace('Store',''))
        No_ratings.append(nrat.text.replace('ratings',''))
        time.sleep(1)
        Rating.append(rat.text)
        Price.append(pri.text.replace('₹',''))
        Return.append(ret.text)
        Expected_delivery.append(exd.text)
        Availability.append(ava.text)
        Other_details.append(oth.text)
    except NoSuchElementException:
        Brand.append('-')
        Product_title.append('-')
        No_ratings.append('-')
        Rating.append('-')
        Price.append('-')
        Return.append('-')
        Expected_delivery.append('-')
        Availability.append('-')
        Other_details.append('-')
    time.sleep(2)
```

```python
gloves = pd.DataFrame({})
gloves['Brand'] = Brand
gloves['Product_title'] = Product_title
gloves['No_ratings'] = No_ratings
gloves['Rating'] = Rating
gloves['Price'] = Price
gloves['Return'] = Return
gloves['Expected_delivery'] = Expected_delivery
gloves['Availability'] = Availability
gloves['Other_details'] = Other_details
gloves['url'] = URL
gloves
```

```python
links = []
url = driver.find_elements_by_xpath('//a[@class="_1fQZEK"]')

for i in url:
    links.append(i.get_attribute('href'))
len(links)
```

```python
brand = []
title = [] 
color = []
ram = [] 
storage = []
primary_cam =  []
secondary_cam = []
display_size = [] 
resolution = []
processor = []
cores = []
battery_cap = []
Price = []

for j in links:
    driver.get(j)
    time.sleep(2)
    
    try:
        tit = driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        pri = driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
        bra = driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[1]/div[1]/div/div[4]/a')
        time.sleep(2)
        title.append(tit.text)
        brand.append(bra.text)
        Price.append(pri.text.replace('₹',''))
    except NoSuchElementException:
        title.append('-')
        brand.append('-')
        Price.append('-')
        
        
    driver.find_element_by_xpath('//button[@class="_2KpZ6l _1FH0tX"]').click()
    time.sleep(1)
    
    other = driver.find_elements_by_xpath('//tr[@class="_1s_Smc row"]')
    
    for det in other:
        text = det.find_element_by_xpath('td[1]').text 
        
        try:
            if text == 'Color':
                color.append(det.find_element_by_xpath('td[2]').text)

            if text == 'RAM':
                ram.append(det.find_element_by_xpath('td[2]').text)

            if text == 'Internal Storage':
                storage.append(det.find_element_by_xpath('td[2]').text)

            if text == 'Primary Camera':
                primary_cam.append(det.find_element_by_xpath('td[2]').text)

            if text == 'Secondary Camera':
                cam = det.find_element_by_xpath('td[2]')
                cam1 = cam.text if cam is not None else '-'
                secondary_cam.append(cam1)                

            if text == 'Display Size':
                display_size.append(det.find_element_by_xpath('td[2]').text)

            if text == 'Resolution':
                resolution.append(det.find_element_by_xpath('td[2]').text)

            if text == 'Processor Type':
                processor.append(det.find_element_by_xpath('td[2]').text)

            if text == 'Processor Core':
                cores.append(det.find_element_by_xpath('td[2]').text)

            if text == 'Battery Capacity':
                battery_cap.append(det.find_element_by_xpath('td[2]').text)
        
        except NoSuchElementException:
            color.append('-')
            ram.append('-')
            storage.append('-')
            primary_cam.append('-')
            secondary_cam.append('-')
            display_size.append('-')
            resolution.append('-')
            processor.append('-')
            cores.append('-')
            battery_cap.append('-')

    
    
```

```python
try:
    driver.find_element_by_xpath('//div[@class="pagination-item"]').click()
    time.sleep(2)
except:
    pass

urls = []
try:
    for _ in range(0,3):
        links = driver.find_elements_by_xpath('//div[@class="title-row"]//a')

        for url in links:
            urls.append(url.get_attribute('href'))
        driver.find_element_by_xpath('//div[@class="pagination-item pagination-next"]').click()
        time.sleep(5)
    len(urls)

except:
    pass
```

```python
links = []
for i in range(0,len(urls),2):
    links.append(urls[i])
links
```

```python
hostel_name = []
ratings = []
total_reviews = []
overall_reviews = []
facilities = []
description = []

for i in links:
    driver.get(i)
    time.sleep(3)
    try:
        name = driver.find_element_by_xpath('//div[@class="headline-container"]//div//h1//div')
        rate = driver.find_element_by_xpath('//div[@class="rating rating-summary-container big clickable"]//div')
        revs = driver.find_element_by_xpath('//div[@class="summary orange"]//div[2]')
        over = driver.find_element_by_xpath('//div[@class="summary orange"]//span')
        faci = driver.find_element_by_xpath('/html/body/div[1]/div/div/div[1]/section/div[11]/div/ul/li[1]/ul')
        desc = driver.find_element_by_xpath('//*[@id="__layout"]/div/div[1]/section/div[6]/div/div[2]/div[2]/div/div[2]')
        
        hostel_name.append(name.text)
        ratings.append(rate.text)
        total_reviews.append(revs.text)
        overall_reviews.append(over.text)
        facilities.append(faci.text)
        description.append(desc.text)
        
    except NoSuchElementException:
        hostel_name.append('-')
        ratings.append('-')
        total_reviews.append('-')
        overall_reviews.append('-')
        facilities.append('-')
        description.append('-')
        time.sleep(3)
        
```
