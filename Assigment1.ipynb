```python

movie_name = []
imdb_rating = []
year_release = []

pages  = np.arange(1,101,50)


for page in pages:
    page = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&view=advanced'+ str(page)+ "&ref_=adv_nxt")
    
    soup = BeautifulSoup(page.text, 'html.parser')
    
    movie_info = soup.find_all('div', class_="lister-item mode-advanced")
    
    for i in movie_info:
        title = movie_name.append(i.h3.a.text.replace('\n',''))
        rate= imdb_rating.append(float(i.strong.text))
        year= year_release.append(i.find('span', class_ ='lister-item-year text-muted unbold').text.strip('()'))  

import pandas as pd
d1 = pd.DataFrame({})
d1['movie_name']= movie_name
d1['imdb_rating']= imdb_rating
d1['year_release'] = year_release
```
